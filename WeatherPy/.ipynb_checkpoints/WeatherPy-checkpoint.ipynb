```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from pprint import pprint

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []
length = [1500]
```

```python
# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=length)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=length)
lat_lngs = zip(lats, lngs)
```

```python
lats
```

```python
lngs
```

```python
# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "metric"

# set up lists to hold reponse info
lat = []
lng = []
temp = []
humidity = []
clouds = []
wind = []
country = []
date = []
```

```python
cities
```

```python
# Build query URL and request your results in Celsius

#uncomment line below to test code
#cities = ["London"]


for city_num in range(len(cities)):
    city = cities[city_num]
    query_url = f"{url}appid={weather_api_key}&q={city}&units={units}"
    
    
    print(f"getting data for {city}, city number {city_num+1} of {len(cities)}")
    try:
        weather_response = requests.get(query_url).json()

        lat.append(weather_response['coord']['lat'])
        lng.append(weather_response['coord']['lon'])
        temp.append(weather_response['main']['temp'])
        humidity.append(weather_response['main']['humidity'])
        clouds.append(weather_response['clouds']['all'])
        wind.append(weather_response['wind']['speed'])
        country.append(weather_response['sys']['country'])
        date.append(weather_response['dt'])
        print("success")
        
    except:
        print(f"Skipping {city} - response found {weather_response}")
        lat.append(np.nan)
        lng.append(np.nan)
        temp.append(np.nan)
        humidity.append(np.nan)
        clouds.append(np.nan)
        wind.append(np.nan)
        country.append(np.nan)
        date.append(np.nan)
    
```

```python
#weather_response  
#create a data frame from lists


meteoroDF = pd.DataFrame({
    "City": cities,
    "Lat": lat,
    "Lng": lng, 
    "Max Temp": temp, 
    "Humidity": humidity, 
    "Cloudiness": clouds, 
    "Wind Speed": wind, 
    "Country": country, 
    "Date": date})
meteoroDF
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
meteoroDF.to_csv('weatherdata.csv', index=False)
meteoroDF
```

## Linear Regression

```python
theNorth = low_humidity.loc[meteoroDF["Lat"] >0, :]
theSouth = low_humidity.loc[meteoroDF["Lat"] <=0, :]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python


## Convert to floats and store Poverty Rate and Bank Count as x and y values
x_values = theNorth['Lat'].astype('float')
y_values = theNorth['Max Temp'].astype('float')

# Run linear regression
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# Plot scatter plot
plt.scatter(x_values,y_values)

# Plot regression line
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")

# Label plot
plt.xlabel('Latitude')
plt.ylabel('Maximum Temperature')
plt.title(f"Graph to regression of Maximum Temperature moving North from the Equator")

# Print r square value
print(f"R squared: {rvalue**2}")

# Show plot
plt.savefig("1.png", format="png")
```
